```python
import pandas as pd
import matplotlib.pyplot as plt


file1 = 'injection_mech_out_20kgs.csv' 
file2 = 'SmearCrack_WJ_out.csv'


data1 = pd.read_csv(file1)
data2 = pd.read_csv(file2)


time1 = data1.iloc[1:555, 0]  # 20000s
time2 = data2.iloc[1:211, 0]  # 20000s




```

```python
J_1 = data1['p1_lagged_perm'].iloc[1:555]  
J_2 = data2['p1_lagged_perm'].iloc[1:211]  
plt.figure(figsize=(12, 6))
plt.plot(time1/60, J_1, label='Case 1', color='black', linestyle='--', linewidth=2.5)
plt.plot(time2/60, J_2, label='Case 2', color='black', linewidth=2.5)

plt.rcParams.update({'font.size': 16})

plt.xlabel('Time (minutes)')
plt.ylabel('Permeability ($m^{2}$)')
plt.title('Change in Permeability with Time')
plt.grid(True)
# plt.legend()
plt.show()
```

```python
J_2 = data1['p1_lagged_poro'].iloc[1:555] 
plt.figure(figsize=(12, 6))
plt.plot(time1/60, J_2, color='orange', linestyle='--', linewidth=2.5)

plt.rcParams.update({'font.size': 16})

plt.xlabel('Time (minutes)')
plt.ylabel('Porosity')
plt.title('Change in Porosity with Time')
plt.grid(True)
# plt.legend()
plt.show()
```

```python
J_3 = data1['p1_vol_strain'].iloc[1:555] 
plt.figure(figsize=(12, 6))
plt.plot(time1/60, J_3, color='black', linestyle='--', linewidth=2.5)

plt.rcParams.update({'font.size': 16})

plt.xlabel('Time (minutes)')
plt.ylabel('Vol_strain')
plt.title('Change in Vol_strain with Time')
plt.grid(True)
# plt.legend()
plt.show()
```
