```python
import pandas as pd
import matplotlib.pyplot as plt


file1 = '2dFrac_1fluid_out_10kgs.csv' 
file2 = '2dFrac_1fluid_out_5kgs.csv'
# file3 = '2dFrac_1fluid_out.csv'

data1 = pd.read_csv(file1)
data2 = pd.read_csv(file2)
# data3 = pd.read_csv(file3)

time1 = data1.iloc[:, 0] 
T_1_10m = data1['j_out']  
T_3_10m = data1['t_out3']

time2 = data2.iloc[:, 0]  
T_1_5m = data2['j_out']  
T_3_5m = data2['t_out3']  

# time3 = data3.iloc[:, 0]  
# T2_1_10m = data3['t_out1']  
# T2_2_10m = data3['t_out3']  

plt.figure(figsize=(12, 6))
plt.plot(time1/(3600*24), T_1_10m, label='20 kg/s', color='blue')
# plt.plot(time1/(3600*24), T_3_10m, label='T3 (10 kg/s)', color='red')
plt.plot(time2/(3600*24), T_1_5m, label='10 kg/s', color='green')
# plt.plot(time2/(3600*24), T_3_5m, label='T3 (5 kg/s)', color='orange')
# plt.plot(time3, T2_1_5m, label='T1', color='black')
# plt.plot(time3, T2_2_5m, label='T3', color='yellow')

plt.xlabel('Time (days)')
plt.ylabel('Energy (J)')
plt.title('Energy_out vs. Time (Zone1_2_3)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
T_1_10 = data1['t_out1']  
T_3_10= data1['t_out3']


time2 = data2.iloc[:, 0]  
T_1_5 = data2['t_out1']  
T_3_5 = data2['t_out3']  


plt.figure(figsize=(12, 6))
plt.plot(time1/(3600*24), T_1_10, label='T2 (20 kg/s)', color='blue', linestyle='--')
plt.plot(time1/(3600*24), T_3_10, label='T1 (20 kg/s)', color='blue')
plt.plot(time2/(3600*24), T_3_5, label='T2 (10 kg/s)', color='green', linestyle='--')
plt.plot(time2/(3600*24), T_1_5, label='T1 (10 kg/s)', color='green')

# plt.plot(time3, T2_1_5m, label='T1', color='black')
# plt.plot(time3, T2_2_5m, label='T3', color='yellow')
plt.xlabel('Time (days)')
plt.ylabel('Temperature')
plt.title('Temperature vs. Time(Zone1_2_3)')
plt.grid(True)
plt.legend()
plt.show()
```
